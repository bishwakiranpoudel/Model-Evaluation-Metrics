# Introduction to Classification Metrics
Classification Metrics are used to evaluate teh performance of classification models, which predict a discrete outcome. These metrics help in assessing how well the model is classifying the outcome into categories

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.datasets import load_iris
```

```python
data = load_iris()
X, y = data.data, data.target

y= (y == 0).astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### Accuracy
Accuracy is the ratio of correctly predicted instances to the total instances.


where TP is true positives, TN is true negatives, FP is false positives and FN is false negatives

#### Use case
Useful as a quick measure, but can be misleading if the data is imbalanced

```python
print(accuracy_score(y_test, y_pred))
```

### Precision
Precision is the ratio of correctly predicted positive observations to the toal predicted positives. It indicates the accuracy of positive predictions.



#### Use Case
Important in scenarios where the cost of false positives is high.

```python
print(precision_score(y_test, y_pred))
```

### Recall (Sensitivity)
Recall is the ratio of correctly predicted positive observations to all observations in the actual class. It measures the ability of the model to identify all relevant instances.



#### Use Case
Crucial in cases where missing a positive case is critical.

```python
print(recall_score(y_test, y_pred))
```

### F1 Score
F1 Score is the harmonic mean of Precision and Recall, providing a balance between the two metrics. It is useful when you need to balance precision and recall.



#### Use Case
Ideal for imbalanced datasets where you need to balance false positives and false negatives.

```python
print(f1_score(y_test, y_pred))
print(model.score(X_test,y_test))
```

### ROC-AUC Curve
Receiver Operating Characteristic curve is a graphical representation of classifier's performance across all classification thresholds. The Area Under the Curve measures the entire two-dimensional area underneath the entire ROC curve.

#### Use Case
AUC provides an aggregate measure of performance across all classification thresholds. A higher AUC indicates a better performing model.

```python
# Calculate ROC-AUC
y_pred_proba = model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
